### Imports
Importing the needed libraries

```python
import os
import cv2 as cv
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import GridSearchCV
from matplotlib import pyplot as plt
from featureextractor import Loader
from weatherclassifier import WeatherClassifier

```

### PATHS & DATASETS
Setting the datasets names, the paths that will be used in all the code and some parameters.

```python
datasets=['MWD','ACDC','UAVid','syndrone']
train_paths=['./datasets/MWD/train',
            './datasets/ACDC/train',
            './datasets/UAVid/train',
            './datasets/syndrone/train']
test_paths=['./datasets/MWD/test',
            './datasets/ACDC/test',
            './datasets/UAVid/test',
            './datasets/syndrone/test']

max_iter=200
random_seed=1
np.random.seed(random_seed)
```

## Classification
Running the classification algorithm on the training sets of all the datasets and testing the performances on the related test data.

#### Only Histograms
Running the classification algorithm using only the full histograms as image features.

```python
num_bins=256


for i in range(len(datasets)):
    print('-------------------------DATASET:{}-------------------------'.format(datasets[i]))
    train_loader=Loader(datasets[i],train_paths[i],num_bins,False,False,False)
    test_loader=Loader(datasets[i],test_paths[i],num_bins,False,False,False)
    
    print('>LOADING TRAINING IMAGES')
    X,y=train_loader.load_images()

    print('>TRAINING CLASSIFIER')
    classifier=WeatherClassifier(X,y,random_seed,max_iter)
    classifier.train_classifier()

    print('>LOADING TEST IMAGES')
    X_test,y_test=test_loader.load_images()

    print('>TESTING')
    score=classifier.score(X_test,y_test)
    print('>SCORES')
    print('score: {}'.format(score))
    print('error: {}'.format(1.0-score))
    print('----------------------------END-----------------------------')
    print()
```

#### Best combination 
Running the algorithm with the best combination of additional parameter (gradient, sharpness and luminance) found and quantized histogram levels.  
We quantize the level axis of the image histogram by using a number of bins lower than the number of levels. By doing this we reduce the number of features representing each image and, thus, possible noise.
We notice that with 100 bins we can have better results in all the datasets.

```python
max_iter=500 #necessary to have a proper convergence
num_bins=100
use_gradient=True
use_sharpness=True
use_luminance=True

for i in range(len(datasets)):
    print('-------------------------DATASET:{}-------------------------'.format(datasets[i]))
    train_loader=Loader(datasets[i],train_paths[i],num_bins,use_gradient,use_sharpness,use_luminance)
    test_loader=Loader(datasets[i],test_paths[i],num_bins,use_gradient,use_sharpness,use_luminance)
    
    print('>LOADING TRAINING IMAGES')
    X,y=train_loader.load_images()

    print('>TRAINING CLASSIFIER')
    classifier=WeatherClassifier(X,y,random_seed,max_iter)
    classifier.train_classifier()

    print('>LOADING TEST IMAGES')
    X_test,y_test=test_loader.load_images()

    print('>TESTING')
    score=classifier.score(X_test,y_test)
    print('>SCORES')
    print('score: {}'.format(score))
    print('error: {}'.format(1.0-score))
    print('----------------------------END-----------------------------')
    print()
```

## Datasets Combination
In the following sections we try to combine the provided datasets in various ways to analyze the behaviour of the proposed method.

#### Syndrone & UAVid
We combine the training and test sets of the Syndrone and the UAVid datasets.


```python
max_iter=200
num_bins=256
use_gradient=True
use_sharpness=True
use_luminance=True

print('-------------------------DATASETS:{}+{}-------------------------'.format(datasets[3],datasets[2]))

syn_train_loader=Loader(datasets[3],train_paths[3],num_bins,use_gradient,use_sharpness,use_luminance)
syn_test_loader=Loader(datasets[3],test_paths[3],num_bins,use_gradient,use_sharpness,use_luminance)

uav_train_loader=Loader(datasets[2],train_paths[2],num_bins,use_gradient,use_sharpness,use_luminance)
uav_test_loader=Loader(datasets[2],test_paths[2],num_bins,use_gradient,use_sharpness,use_luminance)


print('>LOADING TRAINING IMAGES')
X_syn_train,y_syn_train=syn_train_loader.load_images()
X_uav_train,y_uav_train=uav_train_loader.load_images()

X=np.vstack((X_syn_train,X_uav_train))
y=np.append(y_syn_train,y_uav_train)

print('>TRAINING CLASSIFIER')
classifier=WeatherClassifier(X,y,random_seed,max_iter)
classifier.train_classifier()

print('>LOADING TEST IMAGES')
X_syn_test,y_syn_test=syn_test_loader.load_images()
X_uav_test,y_uav_test=uav_test_loader.load_images()

X_test=np.vstack((X_syn_test,X_uav_test))
y_test=np.append(y_syn_test,y_uav_test)

print('>TESTING')
score=classifier.score(X_test,y_test)
print('>SCORES')
print('score: {}'.format(score))
print('error: {}'.format(1.0-score))
print('----------------------------END-----------------------------')
```

#### Total combination
We combine the training and test sets of all the datasets.


```python
max_iter=200
num_bins=256
use_gradient=True
use_sharpness=True
use_luminance=True

print('-------------------------DATASETS:ALL-------------------------')

mwd_train_loader=Loader(datasets[0],train_paths[0],num_bins,use_gradient,use_sharpness,use_luminance)
mwd_test_loader=Loader(datasets[0],test_paths[0],num_bins,use_gradient,use_sharpness,use_luminance)

acdc_train_loader=Loader(datasets[1],train_paths[1],num_bins,use_gradient,use_sharpness,use_luminance)
acdc_test_loader=Loader(datasets[1],test_paths[1],num_bins,use_gradient,use_sharpness,use_luminance)


uav_train_loader=Loader(datasets[2],train_paths[2],num_bins,use_gradient,use_sharpness,use_luminance)
uav_test_loader=Loader(datasets[2],test_paths[2],num_bins,use_gradient,use_sharpness,use_luminance)

syn_train_loader=Loader(datasets[3],train_paths[3],num_bins,use_gradient,use_sharpness,use_luminance)
syn_test_loader=Loader(datasets[3],test_paths[3],num_bins,use_gradient,use_sharpness,use_luminance)


print('>LOADING TRAINING IMAGES')
X_mwd_train,y_mwd_train=mwd_train_loader.load_images()
X_acdc_train,y_acdc_train=acdc_train_loader.load_images()
X_syn_train,y_syn_train=syn_train_loader.load_images()
X_uav_train,y_uav_train=uav_train_loader.load_images()

X=np.vstack((X_mwd_train,X_acdc_train))
y=np.append(y_mwd_train,y_acdc_train)
X=np.vstack((X,X_uav_train))
y=np.append(y,y_uav_train)
X=np.vstack((X,X_syn_train))
y=np.append(y,y_syn_train)

print('total training samples: {}'.format(X.shape[0]))


print('>TRAINING CLASSIFIER')
classifier=WeatherClassifier(X,y,random_seed,max_iter)
classifier.train_classifier()


print('>LOADING TEST IMAGES')

X_mwd_test,y_mwd_test=mwd_test_loader.load_images()
X_acdc_test,y_acdc_test=acdc_test_loader.load_images()
X_syn_test,y_syn_test=syn_test_loader.load_images()
X_uav_test,y_uav_test=uav_test_loader.load_images()

X_test=np.vstack((X_mwd_test,X_acdc_test))
y_test=np.append(y_mwd_test,y_acdc_test)
X_test=np.vstack((X_test,X_uav_test))
y_test=np.append(y_test,y_uav_test)
X_test=np.vstack((X_test,X_syn_test))
y_test=np.append(y_test,y_syn_test)

print('total test samples: {}'.format(X_test.shape[0]))


print('>TESTING')
score=classifier.score(X_test,y_test)
print('>SCORES')
print('score: {}'.format(score))
print('error: {}'.format(1.0-score))
print('----------------------------END-----------------------------')
```
