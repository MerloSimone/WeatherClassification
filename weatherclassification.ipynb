```python
def load_images(path: str, dataset: str,classes: list, format: str):
    if dataset == 'MWD':
        return load_mwd(path,classes,format)
    elif dataset == 'ACDC' :
        return load_other(path,classes,format)
    elif dataset == 'UAVid' :
        return load_other(path,classes,format)
    elif dataset == 'syndrone' :
        return load_other(path,classes,format)


#TBD: da personalizzare i vari metodi per una corretta classificazione.




def load_mwd(path: str, classes: list, format: str):
    X=np.array([])
    y=np.array([])
    
    for cls in classes:
        tmp_path=path+'/'+cls
        files_list = os.listdir(tmp_path)
        for file in files_list:
            if(format in file):
                print(file)
                image=cv.imread(tmp_path+'/'+file)
                if(image is not None) :
                    #COMPUTE FEATURES AND APPEND DATA TO X and y
                    cv.imshow('test0',image)
                    cv.waitKey(1)
                else:
                    print('FALLITO')
                
    return X,y


def load_other(path: str, classes: list, format: str):
    X=np.array([])
    y=np.array([])

    files_list = os.listdir(path)
    for file in files_list:
        print(file)
        image=cv.imread(path+'/'+file)
        if(image is not None) :
            #COMPUTE FEATURES AND APPEND DATA TO X and y 
            cv.imshow('test0',image)
            cv.waitKey(1)
            for cls in classes:
                if( cls in file):
                    print(cls)
                    break #GETTING ONLY THE FIRST MATCHING CLASS
        else:
            print('--------------------------------FALLITO----------------------------------')
        
                
    return X,y
    
```
